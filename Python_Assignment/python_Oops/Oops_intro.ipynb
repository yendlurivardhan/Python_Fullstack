## OOPS : Object oriented programming language

OOPS : OOPS is a programming style/approach that orgaize code around object 


```python
## Creating function (deposite,withdraw,check_balance)

Account_number = "02353423534235676"
Account_holder = "John Doe"
Balance = 5000

# Deposite Function

def deposite(Amount):
    global Balance
    Balance += Amount
    print(f"Deposited: {Amount} and your balance is: {Balance}")

deposite(1000)
```

```python
# Withdraw Function

def withdraw(Amount):
    global Balance
    if Balance > Amount:
        Balance -= Amount
        print(f"withdraw amount: {Amount} and your balance is: {Balance}")
    else:
        print("Insufficient balance")

amount = int(input("Enter Amount: "))
withdraw(amount)
```

```python
# Check balance

def check_balance():
    global Balance
    print(f"Your balance is: {Balance}")

check_balance()
```

```python
# bank account function

class BankAccount():
    # constructor method
    def __init__(self,account_number,account_holder,balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance
    # method inside a class

    def deposite(self,Amount):
        self.balance += Amount
        print(f"Deposited amount: {Amount} and your balance is: {self.balance}")
    
    def withdraw(self,Amount):
        if self.balance > Amount:
            self.balance -= Amount
            print(f"Withdraw amount is: {Amount} and your balance is: {self.balance}")
        else:
            print("Insufficient Balance...")
    def checkBalance():
        print(f"current balance is: {self.balance}")

```
