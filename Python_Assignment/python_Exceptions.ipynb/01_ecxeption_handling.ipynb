```python
try:
    x = 10/"abd"
    print(x)
except Exception as e:
    print(f"Error is {e}")
print("Executed succeessfully")
```

```python
# Handling Multiple Exceotions

try:
    data = [1,2,3]
    index = int(input("Enter index: "))
    print(data[index])
except IndexError:
    print("Error: Please enter valid index")
except ValueError:
    print("Error: Invalid Input please enter number")
print("After executing")
```

```python
try:
    number = int(input("Enter number: "))
    result = 10/number
    print(f"Result : {result}")
except ValueError:
    print("Error: Please enter valid number...")
except ZeroDivisionError:
    print("Error : cannot divide by zero...")
print("Executed successfully...")
```

```python
# Tuple of exceptions 

try:
    value = int(input("Enter a number: "))
    result = 100/value
except (ValueError,ZeroDivisionError) as e:
    print(f"Input or calculation error : {e}")
print("After execution")
```

```python
# Getting exception details 

try:
    file = open("noneexisint.txt", "r")
except FileNotFoundError as e:
    print(f"File error : {e}")
    print(f"Error type : {type(e).__name__}")
```

# The finally clause

 the finally block exceutes regardless of whether an exception occured or not

```python
file = None
try:
    file = open("01_Functions_Assignment.txt","r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not Found")
finally:
    if file:
        file.close()
        print("File closed successfully...")
```

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter first number: "))
    result = num1/num2
except ValueError:
    print("Invalid input: please enter number")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print(f"Result: {result}")
    print("Calculation Successful")
finally:
    print("Calculator Closed")
```
