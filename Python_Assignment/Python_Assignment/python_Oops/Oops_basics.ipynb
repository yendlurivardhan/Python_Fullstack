```python
# class car(display_info,accelerate,break)

class car:
    def __init__(self,brand,model,color):
        self.brand = brand
        self.model = model
        self.color = color
        self.speed = 0

    def display_info(self):
        print(f"Car {self.brand} {self.model}")
        print(f"Car color is {self.color}")
        print(f"Car speed is {self.speed} km/h")
        
    def accelerate(self,increase):
        self.speed += increase
        print(f"Car speed is {self.speed} km/h")

    def brake(self,decrease):
        self.speed -= decrease
        if self.speed < 0:
            self.speed = 0
        print(f"Car speed is {self.speed} km/h")

```

```python
Car1 = car('volvo','XC90','red')
Car2 = car('Toyota','Camry','blue')
```

```python
Car2.display_info()
```

```python
Car2.accelerate(60)
```

```python
Car2.brake(40)
```

```python
# Class Student(add_subject,add_marks,calculate_avg,display_info,check_marks)

class Student:

    def __init__(self,name,roll_number,grade):
        self.name = name
        self.roll_number = roll_number
        self.grade = grade
        self.subjects = []
        self.marks = {}

    def add_subjects(self,subject_name):
        if subject_name not in self.subjects:
            self.subjects.append(subject_name)
            print(f"Subject : {subject_name}") 
        else:
            print(f"Subject {subject_name} already enrolled")

    def add_marks(self,subject_name,Marks):
        if subject_name in self.subjects:
            self.marks[subject_name] = Marks
            print(f"Marks updated : {subject_name} {Marks}")
        else:
            print(f"Student not enrolled subjet {subject_name}")
    
    def calculate_avg(self):
        if len(self.marks) == 0:
            return 0
        total = sum(self.marks.values())
        average = total/len(self.marks)
        return average

    def display_info(self):
        print(f"Student Name: {self.name}")
        print(f"Student Roll Number: {self.roll_number}")
        print(f"Student grade: {self.grade}")
        print(f"Student subjects: {', '.join(self.subjects) if self.subjects else 'None'}")
        if self.marks:
            print("Marks: ")
            for Subject,Marks in self.marks.items():
                print(f"- {Subject} : {Marks}")
            print(f"Average: {self.calculate_avg()}")
        else:
            print("No marks recorded...")

    def check_marks(self):
        avg = self.calculate_avg()
        if avg >= 35:
            print(f"{self.name} is pass")
        else:
            print(f"{self.name} is failed")
```

```python
Student1 = Student("Jonh","002133","A")
```

```python
Student1.add_subjects('science')
Student1.add_subjects('Maths')
Student1.add_subjects('English')
```

```python
Student1.add_marks('science',90)
Student1.add_marks('Maths',92)
Student1.add_marks('English',86)
```

```python
Student1.calculate_avg()
```

```python
Student1.display_info()
```

```python
Student1.check_marks()
```
