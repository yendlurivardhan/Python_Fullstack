```python
# Get even numbers from 1 to 11

even = []
for i in range(1,11):
    if i%2==0:
        even.append(i)
print(even)

# OR

even = [i for i in range(1,11) if i%2==0]
print(even)
```

```python
# Change Lower Case letters to Upper Case letters

names = ["ram","shyam","geetha"]
upper_names = []
for i in names:
    upper_names.append(i.upper())
print(upper_names)

# OR

names = ["ram","shyam","geetha"]
upper_names = [i.upper() for i in names]
print(upper_names)
```

```python
# Get only positive numbers

mixed_numbers = [-3, -1, 0, 2, 5, -8, 10]
for i in mixed_numbers:
    if i>0:
        print(i)
```

```python
# Words with more than 3 letters

words = ["AI", "Python", "ml", "data", "ds"]
for i in words:
    if len(i)>=3:
        print(i)

# OR

word_list = [i for i in words if len(i)>=3]
word_list
```

```python
# Filter Names start with "A"

names = ['Amit', "Raj", "Anita", "Anand", "Bharath"]
for i in names:
    if i.startswith("A"):
        print(i)

# OR

name_list = [i for i in names if i.startswith("A")]
name_list
```

```python
# List of numbers Squaring if positive or Cube if negative

numbers = [2,3,5,-2,-4,-7,9]
result = [i**2 if i>0 else i**3 for i in numbers]
result
```

```python
# If Scores more than or equal to 40 "Pass" else "Fail"

Marks = [40,30,50,60,25]
result = ["Pass" if i>=40 else "Fail" for i in Marks]
result
```

```python
s = "data science is data driven"
words = s.split()
freq = {}
for i in words:
    freq[i] = freq.get(i,0)+1
print(freq)
```
