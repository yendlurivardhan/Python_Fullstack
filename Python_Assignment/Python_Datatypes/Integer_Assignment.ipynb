```python
# Question 1: Calculate the product of first 10 natural numbers using a for loop.
product = 1
for i in range(1,11):
    product *= i
print(f"product of first 10 natural numbers is {product}")
```

```python
# Question 2: Find the remainder when 156 is divided by 7
num = 156
div = 7
remainder = num % div
print(f"The remainder when {num} is divided by {div} is {remainder}")
```

```python
# Question 3: Calculate the square of 4
num = 4
squared = num ** 2
print(f"The square of {num} is {squared}")
```

```python
# Question 4: Find the cube root of 125
num = 125
cube = num ** (1/3)
print(f"Cube of {num} is {cube}")
```

```python
# Question 5: Calculate the sum of digits in number 12345

num = 12345
Sum = 0
temp = num
while num >0:
    value = num % 10
    Sum += value
    num = num//10
print(f" Sum of {temp} is {Sum}")
```

```python

# Question 6: Check if a number is prime or not

def prime(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
num = 29
if prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python

# Question 7: Calculate the factorial of 5

num = 5
factorial = 1
while num>0:
    factorial *=num
    num -= 1
print(f"Factorial of 5 is {factorial}")
```

```python
# Question 8: Calculate the average of numbers: 15, 23, 31, 42, 56

num = [15, 23, 31, 42, 56]
avg = sum(num)/len(num)
print(f"The average of {num} is {avg}")
```

```python

# Question 9: Find the GCD of two numbers 48 and 36
a = 48
b = 36
while b:
    a, b = b, a % b
print(f"The GCD of {48} and {36} is {a}")
```

```python
# Question 10: Calculate the sum of first 20 odd numbers

num = 1
Sum = 0
count = 0
while count<20:
    Sum += num
    num += 2
    count += 1
print(f"The sum of first 20 odd numbers is {Sum}")

```
