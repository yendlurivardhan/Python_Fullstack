```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577
print("Calculate the average of 3.14, 2.718, 1.618, 0.577")
numbers = [3.14, 2.718, 1.618, 0.577]
sum = 0
for i in numbers:
    sum += i
average = sum / len(numbers)
print("The average is:", average)
    
```

```python
# Question 2: Convert 98.6 Fahrenheit to Celsius (F = C * 9/5 + 32)

print("Convert 98.6 Fahrenheit to Celsius")
fahrenheit = 98.6
celsius = (fahrenheit - 32) * 5/9
print("The temperature in Celsius is:", celsius)
```

```python
# Question 3: Calculate the compound interest on $1000 at 5.5% for 3 years

print("Calculate the compound interest on $1000 at 5.5% for 3 years")
principal = 1000
rate = 0.055
time = 3
amount = principal * (1 + rate) ** time
compound_interest = amount - principal
print("The compound interest is:", compound_interest)
```

```python
# Question 4: Find the hypotenuse of a right triangle with sides 3.5 and 4.2

import math
print("Find the hypotenuse of a right triangle with sides 3.5 and 4.2")
side_a = 3.5
side_b = 4.2
hypotenuse = math.sqrt(side_a**2 + side_b**2)
print("The hypotenuse is:", hypotenuse)
```

```python
# Question 5: Calculate the volume of a sphere with radius 7.8

print("Calculate the volume of a sphere with radius 7.8")
radius = 7.8
volume = (4/3) * math.pi * radius**3
print("The volume of the sphere is:", volume)
```

```python
# Question 6: Round 3.14159 to 3 decimal places
number = 3.14159
rounded_number = round(number, 3)
print("3.14159 rounded to 3 decimal places is:", rounded_number)
```

```python
# Question 7: Find the maximum and minimum of the list [2.5, 3.6, 1.2, 4.8, 0.9]

print("Find the maximum and minimum of the list [2.5, 3.6, 1.2, 4.8, 0.9]") 
float_list = [2.5, 3.6, 1.2, 4.8, 0.9]
maximum = max(float_list)
minimum = min(float_list)
print("The maximum is:", maximum)
print("The minimum is:", minimum)
```
