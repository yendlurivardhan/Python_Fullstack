```python
def largest_number(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    elif c>a and c>b:
        return c
    else:
        return "All values are equal"
result = largest_number(3,6,9)
print(f"Largest number is {result}")
```

```python
def prime(number):
    if number<2:
        return "It is not a prime number"
    for i in range(2,int(number**0.5)+1):
        if number%i==0:
            return "It is not a prime number"
    return "It is prime number"
prime(5)
```

```python
def even_odd(number):
    if number%2==0:
        return f"{number} is Even number"
    else:
        return f"{number} is odd number"
even_odd(5)
```

```python
def calculate_mean(*number):
    Sum = 0
    for i in number:
        Sum += i
    avg = Sum/len(number)
    return avg
calculate_mean(1,2,3,4,5)


```

```python
def calculate_median(*number):
    sorted_numbers = sorted(number)
    n = len(number)
    if n%2==0:
        mid1 = sorted_numbers[n//2-1]
        mid2 = sorted_numbers[n//2]
        median = (mid1+mid2)//2
        return median
    else:
        median = sorted_numbers[n//2]
        return median
calculate_median(23,23,23,23,34,43,44,45,45,56,67,67,67,89)
```

```python
def calculate_mode(*number):
    freq = {}
    for i in number:
        freq[i] = freq.get(i,0)+1
    max_count = 0
    for i,j in freq.items():
        if j>max_count:
            max_count = j
    if max_count == 1:
        return "None"
    for i,j in freq.items():
        if j==max_count:
            return i
calculate_mode(1,2,2,3,4)    
```
