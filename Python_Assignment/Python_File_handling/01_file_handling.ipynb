# File Handling


Writing data to a file

Reading data from a file

Updating data inside a file

Deleting a file


    Character Meaning
   ----------------------------------------------------------
    'r'   -   open for reading (default)
    'w'   -   open for writing, truncating the file first
    'x'   -   create a new file and open it for writing
    'a'   -   open for writing, appending to the end of the 
    file if it exists
    'b'   -   binary mode
    't'   -   text mode (default)
    '+'   -   open a disk file for updating (reading and writing)

# Mode 'r' - Read Mode

This is used to read content from the file

```python
with open("note.txt",'r') as f:
    content = f.read()
    print(content)
```

```python
with open("data.txt",'r') as f:
    content = f.read()
    print(content)
```

```python
print(type(content))
```

```python
try:
    with open("note.txt",'r') as  f:
        print(f.read())
except FileNotFoundError:
    print("File doesn't exist")    
```

# Mode "w" - Write (Overwrite)

opens a fiel for write if file doesn't exist, python create it. if it does exist, all previous contents is ERASED and replaced. BE carefull

```python
# Mode - write method

with open("scores.txt",'w') as f:
    f.write("Alice: 95\n")
    f.write("Bob: 90\n")
```

```python
with open("scores.txt",'w') as f:
    f.write("Charle: 94\n")
```

# Mode 'a' - Append Mode

Opens a file for writing bbut movethe cursor to the END of the existing content

```python
# Syntax

# 'a' mode = APPEND (adds to the end , never deletes)
```

```python
with open("attendance.txt", 'a') as f:
    f.write("Day-1: Alice,Bob,charlie\n")
```

```python
# Day 2

with open("attendance.txt",'a') as f:
    f.write("Day-2: Alice,charlie\n")
```

```python
# Day 3

with open("attendance.txt",'a') as f:
    f.write("Day-3:charlie,Bob\n")
```

# 'x'   -   create a new file and open it for writing


```python
try:
    with open("new_report.txt",'x') as f:
        f.write("x - create new file and open it for writing")
    print("File created successfully")
except FileExistsError:
    print("File already exist")
```

```python
try:
    with open("new_report.txt",'x') as f:
        f.write("x - create new file and open it for writing")
        print("File created successfully")
except FileExistsError:
    print("Eror: File already exist")
```
